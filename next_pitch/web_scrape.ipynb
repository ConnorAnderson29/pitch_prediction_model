## Hitter Dataframe Creation(consists of final batting stats from 2018 season)

```python
url ='http://www.espn.com/mlb/stats/batting/_/year/2018/count/41/qualified/false'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

### Populating a dataframe

```python
header = soup.find('tr', attrs={'class': 'colhead'})
```

```python
columns = [col.get_text() for col in header.find_all('td')]
```

```python
final_df = pd.DataFrame(columns=columns)
```

```python
final_df
```

```python
def scrape_espn_player_data(url, final_df):

    '''This functions scrapes ESPN leaderboard data and returns a Pandas Data Frame'''

    for i in range(1,1000, 40):
        url ='http://www.espn.com/mlb/stats/batting/_/year/2018/count/{}/qualified/false'.format(i)
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')

        players = soup.find_all('tr', attrs={'class':re.compile('row player-10-')})
        for player in players:
            # Get Stat for each player
            stats = [stat.get_text() for stat in player.find_all('td')]
            # Create temp dataframe to add later
            temp_df = pd.DataFrame(stats).transpose()
            temp_df.columns = columns
            # Add Data to final_df
            final_df = pd.concat([final_df, temp_df], ignore_index=True)
    return final_df
```

```python
final_df = scrape_espn_player_data('http://www.espn.com/mlb/stats/batting/_/year/2018/count/{}/qualified/false', final_df)
```

```python
final_hitter_df = final_df
```

## Pitcher Dataframe Creation

```python
pitcher_url = 'http://www.espn.com/mlb/stats/pitching/_/year/2018/qualified/false/order/false'
pitch_page = requests.get(pitcher_url)

soup1 = BeautifulSoup(pitch_page.text, 'html.parser')
```

```python
header1 = soup1.find('tr', attrs={'class': 'colhead'})
```

```python
columns1 = [col.get_text() for col in header1.find_all('td')]
```

```python
pitch_final_df = pd.DataFrame(columns=columns1)
pitch_final_df
```

```python
for i in range(1,1000, 40):
    url ='http://www.espn.com/mlb/stats/pitching/_/year/2018/count/{}/qualified/false/order/false'.format(i)
    #url1 = ''
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')

    players = soup.find_all('tr', attrs={'class':re.compile('row player-10-')})
    for player in players:
        # Get Stat for each player
        stats = [stat.get_text() for stat in player.find_all('td')]
        # Create temp dataframe to add later
        temp_df1 = pd.DataFrame(stats).transpose()
        temp_df1.columns = columns1
        # Add Data to final_df
        pitch_final_df = pd.concat([pitch_final_df, temp_df1], ignore_index=True)
pitch_final_df
```

```python
final_hitter_df.to_csv(r"public_data/hitter_data.csv", index=False, sep=',', encoding='utf-8')
```
