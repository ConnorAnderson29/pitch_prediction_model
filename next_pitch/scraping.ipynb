## Dataframe Creation

```python
pitcher_url = 'http://www.espn.com/mlb/stats/pitching/_/count/196/qualified/false/order/false/minip/60'
# 'http://www.espn.com/mlb/stats/pitching/_/year/2018/qualified/false/order/false'
pitch_page = requests.get(pitcher_url)

soup1 = BeautifulSoup(pitch_page.text, 'html.parser')
```

```python
header1 = soup1.find('tr', attrs={'class': 'colhead'})
```

```python
columns1 = [col.get_text() for col in header1.find_all('td')]
```

```python
pitch_final_df = pd.DataFrame(columns=columns1)
pitch_final_df
```

```python
for i in range(1,244, 40):
    url ='http://www.espn.com/mlb/stats/pitching/_/count/{}/qualified/false/order/false/minip/60'.format(i)
    #url1 = ''
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')

    players = soup.find_all('tr', attrs={'class':re.compile('row player-10-')})
    for player in players:
        # Get Stat for each player
        stats = [stat.get_text() for stat in player.find_all('td')]
        # Create temp dataframe to add later
        temp_df1 = pd.DataFrame(stats).transpose()
        temp_df1.columns = columns1
        # Add Data to final_df
        pitch_final_df = pd.concat([pitch_final_df, temp_df1], ignore_index=True)
pitch_final_df
```

```python
final_2019_pdf = pitch_final_df
```

```python
final_2019_pdf = final_2019_pdf.sort_values(by='PLAYER')
```
